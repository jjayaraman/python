```python
import numpy as np

print(np.linspace(0, 1, 5))
print(np.linspace(0, 1, 5, endpoint=False))
print(np.linspace(0, 1, 5, retstep=True))
print(np.random.rand(5))
print(np.random.randn(5))
print(np.random.randint(0, 10, 5))
print(np.random.choice([1, 2, 3, 4, 5], size=3))
print(np.random.choice([1, 2, 3, 4, 5], size=3, replace=False))
print(np.random.choice([1, 2, 3, 4, 5], p=[0.1, 0.2, 0.3, 0.2, 0.2], size=3))
print(np.random.permutation([1, 2, 3, 4, 5]))
print(np.random.shuffle([1, 2, 3, 4, 5]))
```
